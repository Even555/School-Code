```python
# 导入必要的库
import pandas as pd
import matplotlib.pyplot as plt

# 1、导入数据并命名列
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower',
                'weight', 'acceleration', 'year', 'origin', 'name']
df = pd.read_csv('./auto-mpg/auto-mpg.data', delim_whitespace=True, names=column_names)

# 2、执行groupby操作
df_g = df.groupby('cylinders')

# 3、求列'mpg'的均值
Series_df_g = df_g['mpg'].mean()
print(Series_df_g)
# 4、绘制mpg在分组上的均值与cylinder的关系的线图
fig, ax = plt.subplots()
Series_df_g.plot(ax=ax, figsize=(12, 10))
ax.set(title='Relationship between MPG and Cylinder', xlabel='Cylinder', ylabel='MPG')
ax.legend(['Mean MPG'])

# 显示图形
plt.show()
```
