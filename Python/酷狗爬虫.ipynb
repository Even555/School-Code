```python
# coding : UTF-8
#导入程序需要的库
import requests
from bs4 import BeautifulSoup
import time

#通过浏览器获取用户代理
headers ={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0'
}

#定义get_info函数，获取网页信息并输出爬取的内容信息
def get_info(url):
    wb_data = requests.get(url,headers=headers)
    soup = BeautifulSoup(wb_data.text,'lxml')
    ranks = soup.select('span.pc_temp_num')
    titles = soup.select('div.pc_temp_songlist >ul >li > a')
    times = soup.select('span.pc_temp_tips_r > span')
    for rank,title,time in zip(ranks,titles,times):
        data = {
            'rank':rank.get_text().strip(),
            'singer':title.get_text().split(' - ')[1].strip('\n\t'),
            'song':title.get_text().split(' - ')[0].strip('\n\t'),
            'time':time.get_text().strip()
        }
        print(data)

kugou=[r'https://www.kugou.com/yy/rank/home/{}-6666.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-8888.html?from=rank',
       r'https://www.kugou.com/yy/rank/home/{}-59703.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-52144.html?from=rank',
       r'https://www.kugou.com/yy/rank/home/{}-52767.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-24971.html?from=rank',
       r'https://www.kugou.com/yy/rank/home/{}-31308.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-31313.html?from=rank',
       r'https://www.kugou.com/yy/rank/home/{}-54848.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-31310.html?from=rank',
       r'https://www.kugou.com/yy/rank/home/{}-31311.html?from=rank',r'https://www.kugou.com/yy/rank/home/{}-31312.html?from=rank',]
bangdan=["酷狗飙升榜","酷狗TOP500","蜂鸟流行音乐榜","抖音热歌榜","快手热歌榜","DJ热歌榜","内地榜","香港地区榜","台湾地区榜","欧美榜","韩国榜","日本榜"]

#循环调用函数，打印榜单
if __name__ == '__main__':
    for web,bang in zip(kugou,bangdan):
        print(bang)
        if bang == "酷狗TOP500":
            n=24
        else:
            n=6
        urls =[web.format(str(i)) for i in range(1,n)]
        for url in urls:
            get_info(url)
            time.sleep(1)

```
