```python
import jieba
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, accuracy_score
```

```python
train_data = pd.read_csv('data/chnsenticorp/train.tsv', sep='\t')
test_data = pd.read_csv('data/chnsenticorp/test.tsv', sep='\t')
label_map = {0: '负面', 1: '正面'}
x_train, y_train = train_data.text_a.values, train_data.label.values #在问号？处填入正确内容。
x_test, y_test = test_data.text_a.values, test_data.label.values
```

```python
print(train_data, 10)#自己完成代码，目的是训练数据train_data的前10行。
```

```python
class NB_Classifier(object):
    
    def __init__(self):
        # 朴素贝叶斯分类器
        self.model = MultinomialNB( alpha=1) #拉普拉斯平滑：1
        # 使用tf-idf 特征提取
        self.feature_processor = TfidfVectorizer(tokenizer=jieba.cut)
        
    def fit(self, x_train, y_train, x_test, y_test): #在问号？处填入正确内容。
        # tf-idf 特征提取
        x_train_fea = self.feature_processor.fit_transform(x_train)
        self.model.fit(x_train_fea, y_train)
        
        train_accuracy = self.model.score(x_train_fea, y_train)
        print("训练集准确率：{}".format(round(train_accuracy, 3)))
        
        x_test_fea = self.feature_processor.transform(x_test)
        y_predict = self.model.predict(x_test_fea) #提示：把处理好的预测集？？？输入进来进行预测。

        test_accuracy = accuracy_score(y_test, y_predict)
        print("测试集准确率：{}".format(round(test_accuracy, 3)))
        
        y_predict = self.model.predict(x_test_fea)
        print('测试集评估：')
        print(classification_report(y_test, y_predict, target_names=['0', '1']))
        
    def single_predict(self, text):
        text_preprocess = [" ".join(jieba.cut(text))]
        text_fea = self.feature_processor.transform(text_preprocess)
        predict_idx = self.model.predict(text_fea)[0]
        predict_label = label_map[predict_idx]
        predict_prob = self.model.predict_proba(text_fea)[0][predict_idx]
        
        return predict_label, predict_prob
```

```python
nb_classifier = NB_Classifier()
nb_classifier.fit(x_train, y_train, x_test, y_test)
```

```python
nb_classifier.single_predict("外观很漂亮，出人意料地漂亮，做工非常好")
```
